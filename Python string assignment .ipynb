```python
#Q3. How do you access individual characters in a string?

string = "Python"
string_0, string_1, string_2, string_3, string_4, string_5 = string[0],string[1], string[2], string[3], string[-2], string[-1]
#accessing individual character positive index

print(f" This is first character:- {string_0}\n second character:- {string_1}\n Third character:- {string_2}\n Fourth character:- {string_3}")
# accessing individual character negative index

print(f" Fifth character:- {string_4}\n Last character:- {string_5}")
```

```python
#Q4. What method is used to find the length of a string in Python?

# We can find the length of any string with using length()
string = "Python"
string_len = len(string)

print(string_len)
```

```python
#Q5. How can you convert a string to uppercase in Python?
# We can convert any stirng in uppercase with using this upper() method

string = "welcome to my python world"
string_uppercase = string.upper()

print(f"Here converted all string in uppercase:- {string_uppercase}")
```

```python
#Q6. How can you convert a string to lowercase in Python?
# Here we can conver any string in lowercase with using lower() 

string = "WELCOME TO MY PYTHON WORLD"
string_lower = string.lower()

print(f"Here converted everythings in lowercase:- {string_lower}")
```

```python
#Q7. What method is used to replace substrings within a string?
# We can replace string with using using replace()

string = "Welcome to my world"
string_replace = string.replace("world", "Python")

print(f"Here replacing word with Python:- {string_replace}")
```

```python
#Q8. How can you split a string into a list of substrings based on a delimiter?
#we can split string using split() method
string = "Welcome to my python world"
string_split = string.split()

print(f"Here all the strings have been split:- {string_split}")

print(type(string_split))
```
